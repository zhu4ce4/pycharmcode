```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use(['seaborn-talk','dark_background'])
from matplotlib import rcParams
rcParams['font.sans-serif']=['SimHei']
rcParams['axes.unicode_minus']=False
%config InlineBackend.figure_format = 'svg'
```

```python
def read(sheetname):
    i=pd.read_excel('votes.xls',sheet_name=f'{sheetname}')
    i.columns=['排名', '支持票', '名称']
    i['属性']=f'{sheetname}'
    return i
政府=read('政府')
金融=read('金融')
农林=read('农林')
```

```python
data=pd.concat([政府,金融,农林]).set_index(['排名','属性'])
```

```python
data.head()
```

```python
按排名分=data.groupby('属性').mean()
按排名分
```

```python
按属性统计支持票平均=pd.pivot_table(data,index=['属性'],values=['支持票'],aggfunc='mean')
按属性统计支持票平均
#与data.groupby('属性').mean()效果一致
```

```python
plt.plot(按属性统计支持票平均)
```
